````python
import sys
import time
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model
from IPython.display import clear_output
from pathlib import Path
import warnings
from datetime import datetime
import hashlib

# ===========================================================
# ðŸ”§ Logger â€“ zapisuje logi i ostrzeÅ¼enia, bez spamowania konsoli
# ===========================================================
class TeeLogger:
    def __init__(self, filename, silent=True):
        self.log = open(filename, "a", encoding="utf-8")
        self.silent = silent

    def write(self, message):
        if not self.silent:
            sys.__stdout__.write(message)
        self.log.write(message)

    def flush(self):
        self.log.flush()
        if not self.silent:
            sys.__stdout__.flush()

log_file = f"simulation_console_{datetime.now().strftime('%Y%m%d_%H%M%S')}.log"
sys.stdout = TeeLogger(log_file, silent=True)

def custom_warn(message, category, filename, lineno, file=None, line=None):
    sys.stdout.write(warnings.formatwarning(message, category, filename, lineno))
warnings.showwarning = custom_warn

print(f"ðŸ“œ Wszystkie logi i ostrzeÅ¼enia zapisywane w: {log_file}\n")
plt.style.use("seaborn-v0_8-darkgrid")

# ===========================================================
# ðŸ”§ KONFIGURACJA ÅšCIEÅ»KI DO PROJEKTU
# ===========================================================
ROOT = Path.cwd()
while ROOT.name != "ETH-volitality-fotecasting" and ROOT.parent != ROOT:
    ROOT = ROOT.parent
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

try:
    from src.config import LOGS_DIR
except ModuleNotFoundError:
    LOGS_DIR = ROOT / "logs"
    LOGS_DIR.mkdir(exist_ok=True)
    print(f"âš ï¸ Nie znaleziono src.config â€” utworzono lokalny LOGS_DIR â†’ {LOGS_DIR}")

print(f"ðŸ“‚ ÅšcieÅ¼ka projektu dodana do sys.path:\n{ROOT}")

# ===========================================================
# ðŸ” Generator strumieniowy Å›wieczek
# ===========================================================
def stream_candles(csv_path, chunk_size=1000, n_obs=None):
    count = 0
    for chunk in pd.read_csv(csv_path, chunksize=chunk_size, parse_dates=["open_time"]):
        for _, row in chunk.iterrows():
            yield row
            count += 1
            if n_obs and count >= n_obs:
                return

# ===========================================================
# ðŸ§  SYMULACJA EGARCH "LIVE" â€“ wersja ulepszona
# ===========================================================
def simulate_egarch_live_v2(
    csv_path="3000_sample.csv",
    model_filename="egarch_ETH_5m_p1_o1_q1_t.pkl",
    n_obs=3000,
    batch_size=5,
    window_size=600,
    scale_window=400,
    delay=2.0,
    mode_recursive=True
):
    """Symuluje dziaÅ‚anie modelu EGARCH w czasie rzeczywistym â€” Å›wieczka po Å›wieczce."""
    # --- Prosty system nazw logÃ³w ---
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    log_path = LOGS_DIR / f"simulation_logs_{timestamp}.csv"
    result_path = LOGS_DIR / f"simulation_results_{timestamp}.csv"
    print(f"ðŸ§¬ Logi: {log_path.name}")

    # --- Model bazowy ---
    print(f"ðŸ“‚ WczytujÄ™ model: {model_filename}")
    payload = pickle.load(open(LOGS_DIR / model_filename, "rb"))
    fitted, base_scale = payload["model"], payload["scale"]
    model, params, vol = fitted.model, fitted.params, fitted.model.volatility
    prev_params = params.copy()

    returns_memory = list(fitted.model._y[-window_size:])
    forecasts, realized, timestamps, logs = [], [], [], []
    prev_close = None

    print(f"\nðŸš€ Start symulacji (n={n_obs}, batch={batch_size}, window={window_size}, mode={'recursive' if mode_recursive else 'refit'})")

    for i, row in enumerate(stream_candles(csv_path, n_obs=n_obs)):
        current_close = row["close"]

        if prev_close is None:
            prev_close = current_close
            continue

        log_return = np.log(current_close / prev_close)
        prev_close = current_close
        r_std = log_return / base_scale

        returns_memory.append(r_std)
        if len(returns_memory) > window_size:
            returns_memory.pop(0)

        realized.append(abs(log_return))
        timestamps.append(row["open_time"])

        if i % batch_size == 0 and len(returns_memory) > 5:
            subseries = np.array(returns_memory)
            local_window = min(len(subseries), scale_window)
            local_scale = np.std(subseries[-local_window:])
            subseries_std = subseries / local_scale

            try:
                if mode_recursive:
                    fcast_var = fitted.forecast(horizon=1, reindex=False).variance.values[-1, 0]
                    fcast_sigma = np.sqrt(fcast_var) * local_scale * base_scale
                else:
                    temp_model = arch_model(subseries_std, vol="EGARCH",
                                            p=vol.p, o=vol.o, q=vol.q,
                                            dist="t", mean="constant")
                    res = temp_model.fit(disp="off", update_freq=0, starting_values=params)
                    params = res.params
                    fcast_var = res.forecast(horizon=1, reindex=False).variance.values[-1, 0]
                    fcast_sigma = np.sqrt(fcast_var) * local_scale * base_scale
            except Exception as e:
                print(f"âš ï¸ BÅ‚Ä…d przy batchu {i}: {e}")
                params = prev_params.copy()
                continue

            # --- sanity check ---
            if np.any(np.isnan(params)) or np.any(np.abs(params) > 100):
                print(f"âš ï¸ Model siÄ™ rozjechaÅ‚ przy Å›wieczce {i}, przywracam poprzednie parametry.")
                params = prev_params.copy()
                continue
            else:
                prev_params = params.copy()

            # --- smoothing ---
            if forecasts:
                fcast_sigma = 0.9 * forecasts[-1] + 0.1 * fcast_sigma
            fcast_sigma = np.clip(fcast_sigma, 1e-5, 0.05)
            forecasts.append(fcast_sigma)

            # --- monitoring driftu ---
            corr50 = np.corrcoef(forecasts[-50:], realized[-50:])[0, 1] if len(forecasts) > 50 else np.nan

            logs.append({
                "iteration": i,
                "timestamp": str(row["open_time"]),
                "sigma_pred": float(fcast_sigma),
                "local_scale": float(local_scale),
                "corr_rolling50": float(corr50),
                **{f"param_{k}": float(v) for k, v in params.items()}
            })

            if i % 100 == 0:
                pd.DataFrame(logs).to_csv(log_path, index=False)
                print(f"ðŸ’¾ Logi nadpisane ({len(logs)} wpisÃ³w) â†’ {log_path}")

            # --- wizualizacja ---
            clear_output(wait=True)
            x_real = np.arange(len(realized[-300:]))
            x_pred = np.arange(len(forecasts[-300:])) + 1
            plt.figure(figsize=(12, 5))
            plt.plot(x_real, realized[-300:], color="black", alpha=0.6, label="Rzeczywista Ïƒ_t")
            plt.plot(x_pred, forecasts[-300:], color="red", label="Prognozowana Ïƒ_{t+1}")
            plt.legend()
            plt.title(f"EGARCH Live v2 â€“ aktualna Å›wieczka: {row['open_time']}")
            plt.tight_layout()
            plt.show()

        time.sleep(delay)

    if logs:
        pd.DataFrame(logs).to_csv(log_path, index=False)
        print(f"ðŸ§¾ Zapisano koÅ„cowÄ… wersjÄ™ logÃ³w ({len(logs)} wpisÃ³w).")

    result = pd.DataFrame({
        "open_time": timestamps[:len(forecasts)],
        "sigma_pred": forecasts,
        "sigma_real": realized[:len(forecasts)]
    })
    result.to_csv(result_path, index=False)
    print(f"ðŸ’¾ Wyniki zapisane do: {result_path}")
    print(f"\nâœ… Symulacja zakoÅ„czona pomyÅ›lnie ({len(forecasts)} prognoz).")
    return result

````

ðŸ§© 1. Volatility-Conditioned Strategy Layer (Risk Regime Switcher)

ZmiennoÅ›Ä‡ = miara niepewnoÅ›ci â†’ sygnaÅ‚ do przeÅ‚Ä…czania reÅ¼imÃ³w modeli.
W praktyce:

Gdy Ïƒ_pred < dolny kwantyl (np. 30%) â†’ wÅ‚Ä…czasz strategie mean-reversion (bo rynek spokojny, mikro-oscylacje dajÄ… edge).

Gdy Ïƒ_pred > gÃ³rny kwantyl (np. 70%) â†’ wÅ‚Ä…czasz strategie momentum (bo duÅ¼a zmiennoÅ›Ä‡ = breakouty czÄ™stsze niÅ¼ noise).

To daje ci ensemble reÅ¼imowy, ktÃ³ry wÅ‚Ä…cza rÃ³Å¼ne modele wyÅ¼szego rzÄ™du (np. LSTM, RF lub nawet proste filtry sygnaÅ‚Ã³w) w zaleÅ¼noÅ›ci od przewidywanego reÅ¼imu ryzyka.
Jeden EGARCH â†’ dwie lub trzy strategie, selektor decyduje, kto gra.


### ðŸŽ¯ **Idea**

EGARCH nie mÃ³wi ci, *czy cena pÃ³jdzie w gÃ³rÄ™ lub w dÃ³Å‚* â€” on mÃ³wi, **z jakim hukiem** rynek siÄ™ ruszy.
To oznacza, Å¼e w jednym momencie 1 % ruchu jest â€žnormalnyâ€, a w innym 1 % to **katastrofa**.
WiÄ™kszoÅ›Ä‡ modeli ML i statystycznych nie wie o tym â€” dla nich 0.01 to 0.01, zawsze tak samo.
WiÄ™c Å¼eby je uÅ›wiadomiÄ‡, *normalizujesz zwroty przez przewidywanÄ… zmiennoÅ›Ä‡*.

---

### âš™ï¸ **Jak to dziaÅ‚a krok po kroku**

ZaÅ‚Ã³Å¼my, Å¼e masz:

* logarytmiczne zwroty `r_t = log(P_t / P_{t-1})`
* oraz z EGARCH prognozowanÄ… odchyÅ‚kÄ™ standardowÄ… (czyli przewidywanÄ… zmiennoÅ›Ä‡) `Ïƒ_pred_t`.

Tworzysz nowÄ… zmiennÄ…:
[
z_t = \frac{r_t}{Ïƒ_{pred, t}}
]

To jest twoje **volatility-adjusted return** â€“ zwrot przeskalowany ryzykiem.
Ekonomicznie: to jest *ile jednostek zwrotu dostaÅ‚eÅ› za jednostkÄ™ ryzyka*.

---

### ðŸ“ˆ **Intuicja w Å›wiecie algotradingu**

1. Kiedy rynek jest spokojny (`Ïƒ_pred` maÅ‚e), nawet maÅ‚e zwroty sÄ… istotne â†’ po przeskalowaniu rosnÄ… (duÅ¼y z_t).
2. Kiedy rynek jest burzliwy (`Ïƒ_pred` duÅ¼e), ten sam zwrot to tylko szum â†’ po przeskalowaniu maleje (maÅ‚y z_t).

W efekcie twoje dane wejÅ›ciowe (np. do modeli predykcyjnych, feature engineeringu czy strategii) stajÄ… siÄ™:

* *jednostkowo porÃ³wnywalne* niezaleÅ¼nie od pory dnia,
* *stabilniejsze w wariancji* (czyli stacjonarne),
* *pozbawione iluzji mocy* podczas paniki rynkowej.

---

### ðŸ§® **W praktyce**

Wystarczy ci jedna linia w kodzie:

```python
returns_norm = returns / sigma_pred
```

I teraz moÅ¼esz:

* trenowaÄ‡ modele LSTM, RF, XGBoost itp. na `returns_norm` zamiast `returns`,
* obliczaÄ‡ momentum, RSI czy cokolwiek innego na tych znormalizowanych wartoÅ›ciach,
* interpretowaÄ‡ sygnaÅ‚y z innych modeli na wspÃ³lnej â€žskali ryzykaâ€.

---

### ðŸ’¡ **Efekt koÅ„cowy**

TwÃ³j system zaczyna myÅ›leÄ‡ nie w jednostkach ceny, tylko w jednostkach *ryzyka*.
PrzykÅ‚ad:

* normalny model widzi +1 % i myÅ›li: *super, kupuj*.
* model z normalizacjÄ… widzi +1 % przy Ïƒ=0.02 â†’ z_t = 0.5, czyli nic szczegÃ³lnego.
  Ale jeÅ›li Ïƒ=0.005 â†’ z_t = 2 â†’ to rzeczywiÅ›cie silny sygnaÅ‚.

---

### ðŸ”¬ **Dlaczego to dziaÅ‚a**

Bo EGARCH tÅ‚umaczy zmiennoÅ›Ä‡ logarytmicznie (asymetrycznie wzglÄ™dem szokÃ³w),
wiÄ™c `Ïƒ_pred` odzwierciedla nie tylko â€žileâ€, ale teÅ¼ â€žz jakiego rodzaju ruchuâ€ pochodzi ryzyko.
Normalizacja `r_t / Ïƒ_pred_t` usuwa efekt tego nieliniowego napompowania,
co prowadzi do stacjonarnej serii z E[|z_t|] â‰ˆ 1 i Var(z_t) â‰ˆ 1 â€” wymarzonej sytuacji dla dalszego modelowania.

---

Inaczej mÃ³wiÄ…c:
to nie jest kolejny â€žfeatureâ€, tylko **system nerwowy** dla twoich modeli wyÅ¼szego rzÄ™du â€”
pozwala im reagowaÄ‡ na *prawdziwe znaczenie ruchu*, nie na jego goÅ‚y rozmiar.

I wÅ‚aÅ›nie dlatego ta metoda jest elegancka. Zero dodatkowych parametrÃ³w, zero leakage,
a wprowadza adaptacyjnoÅ›Ä‡ do dynamiki rynku bez utraty prostoty.
