### Manhattan Distance

<div style="font-size:20px">
The Manhattan distance between two points  
is the sum of the absolute differences of their coordinates.  
It is also known as City Block distance (like moving along a grid of streets).
</div>

<div style="font-size:22px; margin-top:15px">
$d(A,B)=\sum_{i=1}^{n} |x_i - y_i|$
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def manhattan_distance(a, b):
    a, b = np.array(a), np.array(b)
    return np.sum(np.abs(a - b))

origin = (0, 0)
points = [(2, -1), (3, 0), (2, -2), (2, 3), (0, -1)]

distances = [(p, manhattan_distance(p, origin)) for p in points]
df = pd.DataFrame(distances, columns=["Point", "Manhattan distance"])
display(df)

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = np.abs(X) + np.abs(Y)

levels = [1, 2, 3, 4]
plt.figure(figsize=(7, 7))
contours = plt.contour(X, Y, Z, levels=levels, cmap="Reds")
plt.clabel(contours, inline=True, fontsize=10, fmt="L1=%d")

for (px, py) in points:
    plt.scatter(px, py, s=50, color="blue")
    plt.text(px+0.2, py+0.2, f"{(px,py)}", fontsize=9)

plt.scatter(0, 0, color="black", marker="x", label="Origin")
plt.legend()
plt.gca().set_aspect("equal", adjustable="box")
plt.title("Manhattan distance (2D)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
