### Minkowski Distance (p=3)

<div style="font-size:20px">
The Minkowski distance is a generalization of both Euclidean and Manhattan distances.  
It is defined with a parameter <i>p</i>:  
- For p = 1 → Manhattan distance  
- For p = 2 → Euclidean distance  
- For p = 3 → a smoother shape between diamond and circle
</div>

<div style="font-size:22px; margin-top:15px">
$d(A,B)=\left(\sum_{i=1}^{n} |x_i - y_i|^p \right)^{1/p}$
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def minkowski_distance(a, b, p=3):
    a, b = np.array(a), np.array(b)
    return np.sum(np.abs(a - b) ** p) ** (1/p)

origin = (0, 0)
points = [(2, -1), (3, 0), (2, -2), (2, 3), (0, -1)]

distances = [(pnt, minkowski_distance(pnt, origin, p=3)) for pnt in points]
df = pd.DataFrame(distances, columns=["Point", "Minkowski (p=3)"])
display(df)

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = (np.abs(X) ** 3 + np.abs(Y) ** 3) ** (1/3)

levels = [1, 2, 3, 4]
plt.figure(figsize=(7, 7))
contours = plt.contour(X, Y, Z, levels=levels, cmap="Purples")
plt.clabel(contours, inline=True, fontsize=10, fmt="%.0f")  

for (i, (px, py)) in enumerate(points):
    plt.scatter(px, py, s=50, color="green")
    plt.text(px+0.3, py+0.3, f"{(px,py)}", fontsize=9, color="black")

plt.scatter(0, 0, color="black", marker="x", label="Origin")
plt.legend()
plt.gca().set_aspect("equal", adjustable="box")
plt.title("Minkowski distance (p=3, 2D)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
