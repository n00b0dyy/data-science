### Chebyshev Distance (Lâˆž)

<div style="font-size:20px">
The Chebyshev distance between two points  
is the greatest absolute difference among their coordinates.  
It is also called the maximum metric or chessboard distance  
(because it represents the minimum number of moves a king needs in chess).
</div>

<div style="font-size:22px; margin-top:15px">
$d(A,B)=\max_{i} \, |x_i - y_i|$
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def chebyshev_distance(a, b):
    a, b = np.array(a), np.array(b)
    return np.max(np.abs(a - b))

origin = (0, 0)
points = [(2, -1), (3, 0), (2, -2), (2, 3), (0, -1)]

distances = [(pnt, chebyshev_distance(pnt, origin)) for pnt in points]
df = pd.DataFrame(distances, columns=["Point", "Chebyshev distance"])
display(df)

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = np.maximum(np.abs(X), np.abs(Y))

levels = [1, 2, 3, 4]
plt.figure(figsize=(7, 7))
contours = plt.contour(X, Y, Z, levels=levels, cmap="Greens")


for lvl in levels:
    plt.text(0, lvl, f"{lvl}", fontsize=10, color="black",
             ha="center", va="bottom", weight="bold")

for (px, py) in points:
    plt.scatter(px, py, s=50, color="darkgreen")
    plt.text(px+0.3, py+0.3, f"{(px,py)}", fontsize=9)

plt.scatter(0, 0, color="black", marker="x", label="Origin")
plt.legend()
plt.gca().set_aspect("equal", adjustable="box")
plt.title("Chebyshev distance (Lâˆž, 2D)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
