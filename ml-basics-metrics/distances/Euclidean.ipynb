### Euclidean Distance

<div style="font-size:20px">
The Euclidean distance between two points  
is the "ordinary" straight-line distance in n-dimensional space.  
In 2D it corresponds to the radius of a circle around the origin.
</div>

<div style="font-size:22px; margin-top:15px">
$d(A,B)=\sqrt{\sum_{i=1}^{n}(x_i - y_i)^2}$
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def euclidean_distance(a, b):
    a, b = np.array(a), np.array(b)
    return np.sqrt(np.sum((a - b) ** 2))

origin = (0, 0)
points = [(2, -1), (3, 0), (2, -2), (2, 3), (0, -1)]

distances = [(p, euclidean_distance(p, origin)) for p in points]
df = pd.DataFrame(distances, columns=["Point", "Euclidean distance"])
display(df)

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)

levels = [1, 2, 3, 4]
plt.figure(figsize=(7, 7))
contours = plt.contour(X, Y, Z, levels=levels, cmap="Blues")
plt.clabel(contours, inline=True, fontsize=10, fmt="r=%.0f")

for (px, py) in points:
    plt.scatter(px, py, s=50, color="red")
    plt.text(px+0.2, py+0.2, f"{(px,py)}", fontsize=9)

plt.scatter(0, 0, color="black", marker="x", label="Origin")
plt.legend()
plt.gca().set_aspect("equal", adjustable="box")
plt.title("Euclidean distance (2D)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
