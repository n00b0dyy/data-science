### 1. Cel dowodu

PoC ma udowadniać, że podmiot ( N ) przechowywał niezmienione dane ( D ) w sposób ciągły w przedziale czasowym ([t_0, t_n]).

Formalnie chcemy zdefiniować funkcję weryfikacji
[
\text{Ver}(D, t_0, t_n, \pi) \to {0,1},
]
która zwraca 1 tylko wtedy, gdy dowód (\pi) poświadcza nieprzerwaną egzystencję danych (D) pomiędzy (t_0) a (t_n).

---

### 2. Funkcja ewolucyjna (łańcuch ciągłości)

Każdy węzeł utrzymuje stan haszowy (H_t) definiowany rekurencyjnie:

[
H_t = F(H_{t-1}, , h_t),
]
gdzie
[
h_t = H_{\text{data}}(D ,|, \tau_t),
]
a (\tau_t) to znacznik czasu pochodzący z zaufanego źródła (time oracle).
(F) to funkcja mieszająca — np. hash kryptograficzny z własnością collision-resistance i deterministycznością:
[
F(x,y) = H_{\text{cont}}(x ,|, y).
]

Łańcuch ((H_{t_0}, H_{t_1}, \ldots, H_{t_n})) stanowi nieprzerwany zapis ewolucji stanu.

---

### 3. Dowód ciągłości

Węzeł co pewien okres (\Delta t) publikuje *commitment*
[
C_t = H_t,
]
a cały dowód ma postać
[
\pi = {(C_t, \tau_t)}_{t_0}^{t_n}.
]

Aby dowód był weryfikowalny przy użyciu zk-dowodów lub agregacji hashów, można zdefiniować jego kompresję:
[
\pi' = \text{Compress}(\pi) = \text{zkProve}\big(F, D, [t_0,t_n]\big),
]
tak by weryfikacja była sub-linearna.

---

### 4. Własności, które trzeba udowodnić

1. **Soundness (korektność):**
   Bez znajomości (D) i wszystkich (\tau_t) nie da się skonstruować ważnego (\pi).
   Formalnie:
   [
   \Pr[\text{Ver}(D, t_0, t_n, \pi)=1 ;|; D \text{ nie istnieje w pełni}] \approx 0.
   ]

2. **Temporal consistency (ciągłość czasowa):**
   (\tau_i < \tau_j \Rightarrow C_i ) jest logicznym poprzednikiem (C_j),
   a różnice czasu spełniają (\Delta\tau_j \le T_{\max}).

3. **Efficiency (efektywność):**
   Długość dowodu po kompresji (|\pi'|) jest (O(\log n)) lub (O(1)) dla zakresu (n) commitów.

4. **Consistency / Unicity (jednoznaczność):**
   Nie istnieją dwa różne ciągi (H_t) i (H_t') o tych samych (\tau_t), które przejdą weryfikację:
   [
   \Pr[\text{Ver}(H_t)=\text{Ver}(H_t')=1 \wedge H_t\neq H_t'] \approx 0.
   ]

---

### 5. Funkcje formalne

**Ewolucja:**
[
H_t = H_{\text{cont}}\big(H_{t-1} ,|, H_{\text{data}}(D ,|, \tau_t)\big)
]

**Dowód:**
[
\pi = {(H_t,\tau_t)}_{t=t_0}^{t_n}
]

**Weryfikacja:**
[
\text{Ver}(H_{t_0}, H_{t_n}, \pi) =
\begin{cases}
1, & \text{jeśli } H_{t_i} = F(H_{t_{i-1}}, H_{\text{data}}(D,\tau_i)) ;\forall i,\
0, & \text{w przeciwnym razie.}
\end{cases}
]

**Kompresja (opcjonalna):**
[
\pi' = \text{zkProve}\big(F, D, [t_0,t_n]\big)
\quad\text{tak, że}\quad
\text{Ver}(H_{t_0}, H_{t_n}, \pi') = 1.
]

---

### 6. Integracja z konsensusem

Podczas fazy przejściowej block producer może dostarczać zarówno tradycyjny PoSt jak i PoC.
Mechanizm konsensusu przyjmuje oba typy dowodów, lecz ocenia moc przechowywania w zależności od efektywności (np. PoC → większy weight factor).
